```python
for index, row in soccer.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def contract_details(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# Apply fn to Contract column & create new columns

new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = soccer['Contract'].apply(lambda x: pd.Series(contract_details(x)))

for i in range(len(new_cols)):
    soccer.insert(loc=soccer.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
soccer.sample(2)
```

```python
soccer[['Contract', 'Contract Start', 'Contract End','Contract Length(years)']].sample(2)
```

```python
#Contract categories

def contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
    
#Add contract status column
soccer. insert(soccer.columns.get_loc('Contract Length(years)')+1, 'Contract Status', soccer['Contract'].apply(contract_status))
```

## 13. Height
Player's height in centimeters

```python
soccer['Height'].dtype
```

```python
soccer['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
    
#Apply this fn to the height column
soccer['Height'] = soccer['Height'].apply(convert_height)
soccer['Height'].unique()
```

```python
soccer = soccer.rename(columns={'Height':"Height(cm)"})
soccer.sample(2)
```

## 14. Weight
Player's weight in kilograms

```python
soccer['Weight'].dtype
```

```python
soccer['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)
    
#Apply fn to the weight column
soccer['Weight'] = soccer['Weight'].apply(convert_weight)
soccer['Weight'].unique()
```

```python
soccer = soccer.rename(columns={'Weight':"Weight(kg)"})
soccer.sample(2)
```

## 77. Hits
Number of times the player has been searched for in the FIFA database.

```python
soccer['Hits'].dtype
```

```python
soccer['Hits'].unique()
```

```python
def convert_hits(hits):
    if pd.isna(hits):
        return hits
    elif isinstance(hits, str) and "K" in hits:
        return float(hits.replace("K", "")) * 1000
    else:
        return float(hits)

    
# Apply function to the hits column
soccer['Hits'] = soccer['Hits'].apply(convert_hits)

# Print unique values
print(soccer['Hits'].unique())
```
